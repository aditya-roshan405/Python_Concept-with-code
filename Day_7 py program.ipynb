Problem 3:

```python
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

print(factorial(5))
```

Problem 6:

```python
def is_prime(n):
    if n < 2:
        return False

    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False

    return True

numbers = [2, 3, 4, 5, 6, 7, 8, 9, 10]

primes = list(filter(is_prime, numbers))

print(primes)
```

Problem 9:

```python
def outer():
    x = 10

    def inner():
        nonlocal x
        x += 5
        print(x)

    inner()

outer()
```

Problem 10:

```python
def decorator(func):

    def wrapper():
        print("Starting...")
        func()

    return wrapper

@decorator
def greet():
    print("Hello!")

greet()
```
