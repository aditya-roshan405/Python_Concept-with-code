Program 1:

```python
def fibonacci(n):
    if n <= 1:
        return n
    return fibonacci(n-1) + fibonacci(n-2)

num = 10

for i in range(num):
    print(fibonacci(i), end=" ")
```

Program 2:

```python
memo = {}

def fibonacci(n):
    if n in memo:
        return memo[n]

    if n <= 1:
        return n

    memo[n] = fibonacci(n-1) + fibonacci(n-2)
    return memo[n]

print(fibonacci(10))
```

Program 4:

```python
def armstrong(num):
    digits = str(num)
    power = len(digits)

    total = sum(int(digit) ** power for digit in digits)

    if total == num:
        return True
    else:
        return False

number = 153

if armstrong(number):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
def add_base(base):

    def add(number):
        return base + number

    return add

addition = add_base(10)

print(addition(5))
print(addition(20))
```

Program 7:

```python
def decimal_to_binary(num):

    if num == 0:
        return "0"

    binary = ""

    while num > 0:
        binary = str(num % 2) + binary
        num = num // 2

    return binary

print(decimal_to_binary(25))
```

Program 8:

```python
def even_numbers():

    num = 0

    while True:
        yield num
        num += 2

gen = even_numbers()

for i in range(10):
    print(next(gen))
```

Program 10:

```python
import re

def validate_password(password):

    if len(password) < 8:
        return "Invalid Password"

    if not re.search(r"\d", password):
        return "Invalid Password"

    if not re.search(r"[!@#$%^&*(),.?\":{}|<>]", password):
        return "Invalid Password"

    return "Valid Password"

print(validate_password("Hello@123"))
print(validate_password("hello123"))
```
