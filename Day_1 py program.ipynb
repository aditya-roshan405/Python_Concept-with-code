Program 1:

```python
class Car:
  def __init__(self,company,model,color,year):
    self.color = color
    self.company = company
    self.year = year
    self.model = model

  def disp_details(self):
    print(f"Company: {self.company}")
    print(f"Model: {self.model}")
    print(f"Color: {self.color}")
    print(f"Year: {self.year}")

car1 = Car("Toyota", "Camry", "Platinum White Pearl", 2015)
car1.disp_details()
```

Program 3:

```python
class Car:
    def __init__(self, brand):
        self.brand = brand

class ElectricCar(Car):
    def __init__(self, brand, battery_capacity):
        super().__init__(brand)
        self.battery_capacity = battery_capacity

    def display(self):
        print(f"Brand: {self.brand}")
        print(f"Battery Capacity: {self.battery_capacity} kWh")


e1 = ElectricCar("Tesla", 75)
e1.display()
```

Program 6:

```python
class Student:
    def __init__(self, marks):
        self._marks = marks

    @property
    def marks(self):
        return self._marks

    @marks.setter
    def marks(self, value):
        if value >= 0:
            self._marks = value

s = Student(85)

print(s.marks)

s.marks = 95
print(s.marks)
```

Program 9:

```python
class Demo:
    count = 0

    @classmethod
    def show_count(cls):
        print(cls.count)

    @staticmethod
    def greet():
        print("Hello")

Demo.show_count()
Demo.greet()
```
